```python
import os
import pandas as pd
```

```python
train_path = "../data/Tamil_troll_memes/train_captions.csv"
test_path = "../data/Tamil_troll_memes/test_captions.csv"
test_labels_path = "../data/Tamil_troll_memes/gold_labels_for_test.csv"
```

```python
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```

```python
test_labels = (pd.read_csv(test_labels_path).set_index('ref_caption')['label'] == 'troll').astype('int').to_dict()
#test_labels
```

```python
train_df.head()
```

```python
train_df['split'] = 'train'
train_df['meme_path'] = 'uploaded_tamil_memes/' + train_df['imagename']
train_df['is_troll'] = train_df['imagename'].str.startswith('troll').astype('int')
train_df.head()
```

```python
test_df['split'] = 'test'
test_df['meme_path'] = 'test_img/' + test_df['imagename']
test_df['is_troll'] = test_df['imagename'].apply(lambda x: test_labels[x])
test_df.head()
```

```python
cols = ['split', 'meme_path', 'captions', 'is_troll']
df = pd.concat([train_df[cols], test_df[cols]]).reset_index(drop=True)
df = df.rename(columns={'captions': 'text'})
df['text'] = df['text'].str.lower()
print(df.shape)
df.head()
```

```python
df.tail()
```

```python
df.groupby('split')['is_troll'].value_counts()
```

```python
from indictrans import Transliterator
model = Transliterator(source='eng', target='tam', build_lookup=True)
text = """Sheikh Hamdan bin Zayed Al Nahyan, Ruler's Representative in Al Dhafra Region, attended the opening of the International Atomic Energy Agency's (IAEA) ConvEx-3 'Barakah UAE', which is aimed to test the global emergency response system for a severe accident simulated at the Barakah nuclear power plant.

During the exercise, 75 member states and 12 international organisations are participating to evaluate their emergency response actions in order to identify good practices and highlight areas needing improvement."""
result = model.transform(text)
print(result)
```

```python
df['text_transliterated'] = df['text'].apply(lambda x: model.transform(x))
```
